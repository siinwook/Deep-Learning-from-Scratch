perceptron


```python
def AND(x1,x2):
  w1,w2,theta = 0.5, 0.5, 0.7
  tmp= w1*x1 + w2*x2
  if tmp<=theta:
    return 0
  else:
    return 1
```

```python
for x1,x2 in [(0,0),(0,1),(1,0),(1,1)]:
  print(f"{x1} & {x2} : {AND(x1,x2)}")
```

logic gate with perceptron

```python
# AND, NAND, OR는 같은 퍼셉트론 구조, 가중치에만 차이
def AND(x1,x2):
  x=np.array([x1,x2])
  w=np.array([0.5,0.5])
  b=-0.7
  tmp = np.sum(w*x)+b
  if tmp<=0:
    return 0
  else:
    return 1

def NAND(x1,x2):
    x=np.array([x1,x2])
    w=np.array([-0.5,-0.5])
    b=0.7
    tmp = np.sum(w*x)+b
    if tmp<=0:
      return 0
    else:
      return 1

def OR(x1,x2):
    x=np.array([x1,x2])
    w=np.array([0.5,0.5])
    b=-0.2
    tmp = np.sum(w*x)+b
    if tmp<=0:
      return 0
    else:
      return 1
```

```python
for x1,x2 in [(0,0),(0,1),(1,0),(1,1)]:
  print(f"{x1} and {x2} : {AND(x1,x2)}")
print("")
for x1,x2 in [(0,0),(0,1),(1,0),(1,1)]:
  print(f"{x1} nand {x2} : {NAND(x1,x2)}")
print("")
for x1,x2 in [(0,0),(0,1),(1,0),(1,1)]:
  print(f"{x1} or {x2} : {OR(x1,x2)}")
```

nonlinear gate with 2-layer perceptron

```python
def XOR(x1,x2): #2층 퍼셉트론, 비선형.
  s1=OR(x1,x2)
  s2=NAND(x1,x2)
  y=AND(s1,s2)
  return y
```

```python
for x1,x2 in [(0,0),(0,1),(1,0),(1,1)]:
  print(f"{x1} xor {x2} : {XOR(x1,x2)}")
```
